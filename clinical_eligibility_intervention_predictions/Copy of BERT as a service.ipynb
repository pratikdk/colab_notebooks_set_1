```
!ls
#!rm -rf ml_datasets/ 
!ls ml_datasets/ 
import zipfile
zf = zipfile.ZipFile("ml_datasets/labeledEligibilitySample1000000.zip") 
data = pd.read_csv(zf.open('labeledEligibilitySample1000000.csv'))
```

```
data = data.iloc[:,[1,2]]
data.loc[data.shape[0],:]=data.columns
data.columns=['label','clinical_statements']
# Lower Case
data['clinical_statements'] =data['clinical_statements'].str.lower()

data['clinical_statements']  = [x.translate(str.maketrans('', '', string.punctuation)) for x in data['clinical_statements']]
data['clinical_statements'] = [x.translate(str.maketrans('', '', string.digits)) for x in data['clinical_statements']]

data['clinical_statements'] = [x.replace('and','') for x in data['clinical_statements']]
data['clinical_statements']=[' '.join(x.split()) for x in data['clinical_statements']]
```

```
data = data.sample(n= int(np.round(data.shape[0]*0.1)))
```

```
# CNN Layer architecture
print('Training model.')

# define the model
model = Sequential()
model.add(Input(shape=(768,)))
model.add(Conv1D(filters = 128,kernel_size=5, strides=1,activation='relu'))
model.add(MaxPooling1D(pool_size=5,strides=1))
model.add(Conv1D(filters = 128,kernel_size=5, strides=1,activation='relu'))
model.add(MaxPooling1D(pool_size=5,strides=1))
model.add(Conv1D(filters = 128,kernel_size=5, strides=1,activation='relu'))
model.add(MaxPooling1D(pool_size=35,strides=1))
model.add(GlobalMaxPooling1D())
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='softmax'))

# compile the model
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['acc'])

# summarize the model
print(model.summary())


```

```
import json
import requests
def get_embeddings(texts):
    headers = {
        'content-type':'application/json'
    }
    data = {
        "id":123,
        "texts":texts,
        "is_tokenized": False
    }
    data = json.dumps(data)
    r = requests.post("http://localhost:" + port_num + "/encode", data=data, headers=headers).json()
    return r['result']
```
